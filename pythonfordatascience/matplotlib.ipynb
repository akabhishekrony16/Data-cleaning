# setting the title,axis,labels,ticks and ticktables

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
```

```python
#To change the X axis ticks,its easiest to use set_xticks and set_xticklabels.The former gives position of the ticks 
#along the data range;by default these locations will also be the labels.But we can set any other values as the labels
#using set_xticklabels:
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],rotation =30,fontsize='small')
```

```python
#lastly set_xlabel gives a aname to the x axis and set_title the subplot title
```

```python
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

```python
fig
```

# Bar plots

```python
#for verticals bars pass kind=bar and for horizontal pass kind =barh
```

```python
fig,axes = plt.subplots(2,1)
data =np.abs(Series(np.random.randn(16),index=list('abcdefghijklmnop')))
data.plot(kind='bar',ax=axes[0],color='k',alpha=0.7)
data.plot(kind='barh',ax=axes[1],color='k',alpha=0.7)
```
